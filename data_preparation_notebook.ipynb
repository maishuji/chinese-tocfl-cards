# Data Analysis

```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt

excel_path = Path("chinese_8000_words_tocfl", "chinese_8000_tocfl_20240923.xlsx")
print(excel_path)
```

```python
sheets = pd.read_excel(excel_path, sheet_name=None, engine="openpyxl")

selected_sheets = ["準備級一級(Novice 1)",
                   "準備級二級(Novice 2)",
                   "入門級(Level 1)",
                   "基礎級(Level 2)",
                   "進階級(Level 3)",
                   "高階級(Level 4)",
                   "流利級(Level 5)",
                   ]
sheets = {name: sheets[name] for name in selected_sheets}
```

```python
for sheet_name, df in sheets.items():
    print(f"Sheet name: {sheet_name}, number of rows: {len(df)}")
    print("Header columns:\n\t", df.columns.tolist())
```

```python
out_dir = Path("data", "csv")
out_dir.mkdir(parents=True, exist_ok=True)
```

```python
def normalize_df(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    # propagate merged-cell values (common in Excel vocab sheets)
    df = df.ffill(axis=0)
    # remove fully-empty columns
    df = df.dropna(axis=1, how="all")
    # normalize column names (strip, lowercase)
    df.columns = [str(c).strip() for c in df.columns]
    # mapping examples: extend as needed for your workbook
    mapping = {
        "任務領域\nContext": "Context", "詞彙\nVocabulary": "Vocabulary", 
        "漢語拼音\nPinyin": "Pinyin",
        "詞類\nParts of Speech": "Parts of Speech"
    }
    # rename if keys exist
    rename_map = {col: mapping[col] for col in df.columns if col in mapping}
    if rename_map:
        df = df.rename(columns=rename_map)
    return df
```

```python
for sheet_name, df in sheets.items():
    df_norm = normalize_df(df)
    out_path = out_dir / f"{sheet_name}.csv"
    df_norm.to_csv(out_path, index=False, encoding="utf-8-sig")
    print(f"Saved {out_path}  rows={len(df_norm)} cols={len(df_norm.columns)}")
    display(df_norm.head())
```
