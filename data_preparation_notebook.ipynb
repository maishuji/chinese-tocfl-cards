```python
out_dir = Path("data", "csv")
out_dir.mkdir(parents=True, exist_ok=True)
```

```python
def normalize_df(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    # propagate merged-cell values (common in Excel vocab sheets)
    df = df.ffill(axis=0)
    # remove fully-empty columns
    df = df.dropna(axis=1, how="all")
    # normalize column names (strip, lowercase)
    df.columns = [str(c).strip() for c in df.columns]
    # mapping examples: extend as needed for your workbook
    mapping = {
        "任務領域\nContext": "Context", "詞彙\nVocabulary": "Vocabulary", 
        "漢語拼音\nPinyin": "Pinyin",
        "詞類\nParts of Speech": "Parts of Speech"
    }
    # rename if keys exist
    rename_map = {col: mapping[col] for col in df.columns if col in mapping}
    if rename_map:
        df = df.rename(columns=rename_map)
    return df
```

```python
for sheet_name, df in sheets.items():
    df_norm = normalize_df(df)
    out_path = out_dir / f"{sheet_name}.csv"
    df_norm.to_csv(out_path, index=False, encoding="utf-8-sig")
    print(f"Saved {out_path}  rows={len(df_norm)} cols={len(df_norm.columns)}")
    display(df_norm.head())
```

```python
# Translation helper cell
# This cell will: detect the headword column, translate values to English,
# and write translated CSVs into `data/csv_translated/`.
from pathlib import Path
import pandas as pd

repo = Path("/home/maishuji/Workplace/chinese-flash-cards")
csv_dir =  Path("data", "csv")
out_dir = Path("data", "csv_translated")
out_dir.mkdir(parents=True, exist_ok=True)

# try to import a MarianMT model for zh->en translation
HAS_MARIAN = False
try:
    from transformers import MarianMTModel, MarianTokenizer
    HAS_MARIAN = True
except Exception as e:
    print("MarianMT not available locally (will try googletrans fallback).")
    print(e)

# fallback to googletrans if Marian not installed
HAS_GOOGLETRANS = False
if not HAS_MARIAN:
    try:
        from googletrans import Translator
        gt = Translator()
        HAS_GOOGLETRANS = True
    except Exception as e:
        print("googletrans not available. Install transformers or googletrans before running this cell.")
        print(e)


```

```python
def detect_headword_column(cols):
    # common candidate column names (add more if needed)
    candidates = ['headword','Vocabulary','詞彙','漢字','詞','word','詞語','詞目']
    for c in cols:
        if c in candidates:
            return c
    # fallback: prefer the first column that contains Chinese characters
    import re
    zh_re = re.compile('[一-鿿]')
    for c in cols:
        try:
            sample = str(cols[c]) if isinstance(cols, dict) else ''
        except Exception:
            sample = ''
    # if above heuristic fails, just pick the first column
    return None

def translate_texts_marian(texts, model_name='Helsinki-NLP/opus-mt-zh-en'):
    from transformers import MarianMTModel, MarianTokenizer
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)
    batch = 16
    out = []
    for i in range(0, len(texts), batch):
        batch_texts = texts[i:i+batch]
        inputs = tokenizer(batch_texts, return_tensors='pt', padding=True, truncation=True)
        translated = model.generate(**inputs)
        dec = tokenizer.batch_decode(translated, skip_special_tokens=True)
        out.extend(dec)
    return out

def translate_texts_google(texts):
    from googletrans import Translator
    translator = Translator()
    out = []
    for t in texts:
        if not t or str(t).strip()=='' :
            out.append('')
            continue
        res = translator.translate(str(t), src='zh-cn', dest='en')
        out.append(res.text)
    return out
```

```python
# Process each CSV in the folder
for p in sorted(csv_dir.glob('*.csv')):
    print('Processing', p.name)
    df = pd.read_csv(p, encoding='utf-8-sig')
    cols = df.columns.tolist()
    # try to detect headword column by common names
    head_candidates = ['headword','Vocabulary','詞彙','漢字','詞','word','詞語','詞目']
    head_col = None
    for c in head_candidates:
        if c in cols:
            head_col = c
            break
    if head_col is None:
        # heuristic: first column that contains any CJK ideograph in its first 50 non-null values
        import re
        zh_re = re.compile('\一-\鿿')
        for c in cols:
            sample_vals = df[c].dropna().astype(str).head(50).tolist()
            if any(zh_re.search(str(v)) for v in sample_vals):
                head_col = c
                break
    if head_col is None:
        print('Could not detect headword column for', p.name, '- skipping')
        continue
    print('Detected headword column:', head_col)
    texts = df[head_col].fillna('').astype(str).tolist()
    if HAS_MARIAN:
        try:
            translations = translate_texts_marian(texts)
        except Exception as e:
            print('Marian translation failed:', e)
            translations = []
    elif HAS_GOOGLETRANS:
        translations = translate_texts_google(texts)
    else:
        raise RuntimeError('No translation backend available. Install `transformers` + model or `googletrans`.')

    # attach translations and save
    df['Translation'] = translations
    out_path = out_dir / p.name
    df.to_csv(out_path, index=False, encoding='utf-8-sig')
    print('Wrote', out_path)
```

## Merging into a single csv

```python
# Merge all translated CSVs into one, with an additional column for the source level named "Level"
import pandas as pd
from pathlib import Path
csv_translated_dir = Path("data", "csv_translated")
all_dfs = []
for p in sorted(csv_translated_dir.glob('*.csv')):
    level_name = p.stem  # use the file name (without extension) as level name
    df = pd.read_csv(p, encoding='utf-8-sig')
    df['Level'] = level_name
    all_dfs.append(df)
merged_df = pd.concat(all_dfs, ignore_index=True)
out_path = Path("data", "chinese_8000_tocfl_translated_merged.csv")
merged_df.to_csv(out_path, index=False, encoding='utf-8-sig')
print(f"Merged translated CSVs written to {out_path}, total rows: {len(merged_df)}")
```
